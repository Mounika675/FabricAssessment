```python
%%sql
create view TEMP1 as (
SELECT product_name,avg(price) as price FROM Lakehouse1.erp_table  where price is not null group by product_name)
```

```python


df = spark.sql("SELECT product_id, \
        erp.product_name,\
case when category is null then 'Category not Available'\
else Category end as Category,\
coalesce(erp.price,t.price) as Price,\
erp.stock_quantity\
 from Lakehouse1.erp_table erp join temp1 t on erp.product_name=t.product_name")
display(df)
df.overwrite.format("delta").saveAsTable("New_ERP_data")
```
